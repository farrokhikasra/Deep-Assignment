```python
from torchvision.datasets import SVHN, MNIST
import torchvision
import numpy as np
import torch
import matplotlib.pyplot as plt
import torch.nn as nn
from torchvision import transforms
```

```python
# Initialize train and test datasets
train_set = MNIST('../data',
                           train=True,
                           download=True,
                           transform=transforms.ToTensor())
test_set = MNIST('../data',
                          train=False,
                          download=True,
                          transform=transforms.ToTensor())

# Initialize train and test data loaders
train_mnist = torch.utils.data.DataLoader(train_set,
                                           batch_size=128,
                                           shuffle=True,
                                           drop_last=True)
test_mnist = torch.utils.data.DataLoader(test_set,
                                          batch_size=128,
                                          shuffle=True,
                                          drop_last=True)
```

```python
train_set1 = SVHN(root='..data/', download=True, split='train',
                   transform=torchvision.transforms.Compose([transforms.ToTensor(), 
                                                             transforms.Grayscale(num_output_channels=1)]),)

test_set1 = SVHN(root='..data/', download=True, split='test',
                   transform=torchvision.transforms.Compose([transforms.ToTensor(), 
                                                             transforms.Grayscale(num_output_channels=1)]),)

train_svhn = torch.utils.data.DataLoader(train_set1, 
                                           batch_size=128,
                                           shuffle=True,
                                           drop_last=True)
test_svhn = torch.utils.data.DataLoader(test_set1,
                                          batch_size=128,
                                          shuffle=True,
                                          drop_last=True)
```

```python
feature_model = nn.Sequential(nn.Conv2d(1, 32, 5), nn.BatchNorm2d(32), nn.ReLU(), 
                      nn.MaxPool2d(2, stride=2),
                      nn.Conv2d(32, 64, 5), nn.BatchNorm2d(64), nn.ReLU(),
                      nn.Conv2d(64, 64, 3), nn.BatchNorm2d(64), nn.ReLU(),
                      nn.AdaptiveAvgPool2d((1,1)), nn.Flatten())
```

```python
import torch
from torch.utils import data
import torch.nn.functional as F

def learning(train_dataloader, test_dataloader, model):

    # Training hyper parameters
    train_batch_size = 128
    n_epochs = 20

    train_losses = [] # use to append the avg loss for each minibatch 
    train_accs = [] # use to append the avg acc of minibatch
    
    optimizer = torch.optim.Adam(model.parameters(),lr=1e-3)

    #iterate over the minibatches
    for epoch in range(n_epochs):
        for data,target in train_dataloader:

            #compute model output
            output = model(data)
            #compute the gradient for the minibatch
            loss = F.cross_entropy(output,target)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            # track training loss and accuracy
            train_losses.append(loss.item())
            prediction = output.argmax(1)
            train_accs.append(sum(prediction == target).item()/train_batch_size)
            
        # print training progress
        print(f' epoch {epoch}: train loss {sum(train_losses)/(len(train_dataloader)*(epoch+1)):.4f}- train accuracy {sum(train_accs)/(len(train_dataloader)*(epoch+1))*100:.2f}%')
    #Plot the train loss and acc
    fig,axs = plt.subplots(1,2,figsize=(20,10))
    axs[0].plot(train_losses)
    axs[0].set_title('train loss')
    axs[0].set_xlabel('iteration')
    axs[0].set_ylabel('loss')
    axs[1].plot(train_accs)
    axs[1].set_title('train accuracy')
    axs[1].set_xlabel('iteration')
    axs[1].set_ylabel('accuracy')
 
    # save train loss and accuracy for further comparison
    train_loss_sgd = train_losses
    train_accs_sgd = train_accs
    
    loss = 0
    accuracy = 0

    for data, target in test_dataloader:

        #compute model output
        output = model(data)
        #comptue accuracy for minibatch
        prediction = output.argmax(1)
        accuracy += sum(prediction == target).item()/128
        #compute loss for minibatch
        loss += F.cross_entropy(output,target).item()
    #aggregate loss and accuracy for all test data
    print(f'test loss: {loss/len(test_dataloader):4f}')
    print(f'test accuracy: {accuracy/len(test_dataloader)*100:.2f}%')
```

```python
learning(train_mnist, test_mnist, feature_model)
learning(train_svhn, test_svhn, feature_model)
```

```python
mnist_head = nn.Sequential(nn.Linear(64, 64), nn.ReLU(), nn.Linear(64, 10))
svhn_head = nn.Sequential(nn.Linear(64, 64), nn.ReLU(), nn.Linear(64, 10))
```

```python
mnist_model = nn.Sequential(feature_model, mnist_head)
svhn_model = nn.Sequential(feature_model, svhn_head)
```

```python
def multitask_learning(mnist_train, mnist_test, svhn_train, svhn_test, mnist_model, svhn_model):

    # Training hyper parameters
    train_batch_size = 128
    n_epochs = 20
    
    train_losses = [] # use to append the avg loss for each minibatch 
    mnist_train_accs = [] # use to append the avg acc of minibatch
    svhn_train_accs = []
    parameters = list(mnist_model.parameters()) + list(svhn_model.parameters())
    optimizer = torch.optim.Adam(parameters,lr=1e-3)

    #iterate over the minibatches
    for epoch in range(n_epochs):
        svhn_iter = iter(svhn_train)
        for data1, target1 in mnist_train:
            output1 = mnist_model(data1)
            loss1 = F.cross_entropy(output1, target1)
            prediction1 = output1.argmax(1)
            mnist_train_accs.append(sum(prediction1 == target1).item() / train_batch_size)
            
            data2, target2 = next(svhn_iter)
            output2 = svhn_model(data2)
            loss2 = F.cross_entropy(output2, target2)
            prediction2 = output2.argmax(1)
            svhn_train_accs.append(sum(prediction2 == target2).item() / train_batch_size)
            loss = loss1 + loss2
            train_losses.append(loss.item())
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        # print training progress
        print(f' epoch {epoch}: train loss {sum(train_losses)/((len(mnist_train) + len(svhn_train))*(epoch+1)):.4f}- MNIST train accuracy {sum(mnist_train_accs)/(len(mnist_train)*(epoch+1))*100:.2f}%- SVHN train accuracy {sum(svhn_train_accs)/(len(svhn_train)*(epoch+1))*100:.2f}%')
    #Plot the train loss and acc
    fig,axs = plt.subplots(1,3,figsize=(20,10))
    axs[0].plot(train_losses)
    axs[0].set_title('train loss')
    axs[0].set_xlabel('iteration')
    axs[0].set_ylabel('loss')
    
    axs[1].plot(mnist_train_accs)
    axs[1].set_title('MNIST train accuracy')
    axs[1].set_xlabel('iteration')
    axs[1].set_ylabel('MNIST accuracy')
    
    axs[2].plot(svhn_train_accs)
    axs[2].set_title('SVHN train accuracy')
    axs[2].set_xlabel('iteration')
    axs[2].set_ylabel('SVHN accuracy')
 
    loss = 0
    mnist_accuracy = 0
    svhn_accuracy = 0
    for data, target in mnist_test:
        #compute model output
        output = mnist_model(data)
        #compute accuracy for minibatch
        prediction = output.argmax(1)
        mnist_accuracy += sum(prediction == target).item() / 128
        #compute loss for minibatch
        loss += F.cross_entropy(output,target).item()
    #aggregate loss and accuracy for all test data
    print(f'MNIST test accuracy: {mnist_accuracy / len(mnist_test)*100:.2f}%')
    
    for data, target in svhn_test:
        #compute model output
        output = svhn_model(data)
        #comptue accuracy for minibatch
        prediction = output.argmax(1)
        svhn_accuracy += sum(prediction == target).item() / 128
        #compute loss for minibatch
        loss += F.cross_entropy(output,target).item()
    #aggregate loss and accuracy for all test data
    print(f'SVHN test accuracy: {svhn_accuracy / len(svhn_test)*100:.2f}%')
    print(f'test loss: {loss/(len(mnist_train) + len(svhn_train)):4f}')
```

```python
multitask_learning(train_mnist, test_mnist, train_svhn, test_svhn, mnist_model, svhn_model)
```
